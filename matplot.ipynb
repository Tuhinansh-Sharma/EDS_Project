Line graph


```python
import matplotlib.pyplot as plt
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
plt.plot(x_values,y_values)

#coloring the line
plt.plot(x_values,y_values,marker='o',color='r')


#setting x and y axis range
plt.ylim(1,30)
plt.xlim(1,5)


#naming the x an y axis
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.show()
```

Bar graph


```python
import matplotlib.pyplot as plt
#create data for plotting
values=[5,6,3,7,2]
names=["a","b","c","d","e"]
plt.bar(names,values,color="yellowgreen")#'h' flips the graph in barh
plt.show()
```

Histogram

```python
ages=[2,5,70,40,30,45,50,45,43,40,44,60,7,81,15,17,10,11,13,23,24,36,76,20,40]
range=(0,100)
bins=10
plt.hist(ages,bins,range,color='green',histtype='bar',rwidth=0.8)
plt.xlabel('ages')
plt.ylabel('No. of people')
plt.title('My histogram')
plt.show()
```

Scatter plots

```python
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
plt.scatter(x_values, y_values,s=30,color='pink')
plt.show()
```

Pie Chart


```python
activites=['eat','sleep','work','play']
slices=[3,7,8,6]
colors=['r','y','g','b']
plt.pie(slices,labels=activites, colors=colors, startangle=90, shadow=True, explode=(0,0,0.1,0), radius=1.2)
plt.legend()
plt.show()
```

Density plot


```python
import numpy as np
data=np.random.randn(1000)
plt.hist(data,density=True,bins=30,alpha=0.5)
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Density plot')
plt.show()
```

Facet grid


```python
import seaborn as sns
titanic=sns.load_dataset('titanic')
grid=sns.FacetGrid(titanic,row='class',col='sex')
grid.map(sns.histplot,'age')
grid.set_axis_labels('Age','Count')
grid.set_titles(row_template='{row_name}Class',col_template='{col_name}Class')
plt.show()
```

Group plots

```python
import numpy as np
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
fig,axes=plt.subplots(3,1,figsize=(8,10))
axes[0].plot(x,y1,color='red')
axes[0].set_title('Plot 1: Sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,y2,color='blue')
axes[1].set_title('Plot 2: Cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

axes[2].plot(x,y3,color='green')
axes[2].set_title('Plot 3: Tan(x)')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
plt.show()
```

Panels

```python
import plotly.graph_objects as go
import plotly.subplots as sp
x=[1,2,3,4,5];y1=[1,4,9,16,25];y2=[10,8,6,4,2];y3=[1,8,27,64,125]
fig=sp.make_subplots(rows=1,cols=3,subplot_titles=('Panel 1','Panel 2','Panel 3'))
fig.add_trace(go.Scatter(x=x,y=y1,name='Trace 1'),row=1,col=1)
fig.add_trace(go.Scatter(x=x,y=y2,name='Trace 2'),row=1,col=2)
fig.add_trace(go.Scatter(x=x,y=y3,name='Trace 3'),row=1,col=3)
fig.update_layout(showlegend=True)
fig.show()


```
